### Task (b) 

<font size="3">Find |V |, |E|,  (G), (G), rad(G), diam(G), girth(G), center(G), (G), (G).</font>  

<font size="3">We use standard functions of **network x** to find respective values </font>  

````python
G_c = G.subgraph(max(nx.connected_components(G))) #finding the the largest connected component of the initial graph
````

````python
plt.rcParams["figure.figsize"] = (20, 10)
nx.draw_planar(G_c, with_labels = True, node_size = 100, node_color = 'lawngreen', edge_color = 'skyblue', font_size = 12)
# let's just plot it 
````

````python
import numpy as np

print('|V|:', G_c.number_of_nodes())
print('|E|:', G_c.number_of_edges())

#print(G_c.degree())
degrees = np.array([i[1] for i in G_c.degree()])
min_degree = np.min(degrees)
max_degree = np.max(degrees)

print('Minimal graph degree: ', min_degree)
print('Maximal graph degree: ', max_degree)
````

````python
print('Graph diameter: ', nx.diameter(G_c))

print('Graph girth: ', min([len(cycle) for cycle in nx.cycle_basis(G_c)]))

print('Graph center: ', nx.center(G_c))
print('Minimum vertex cut: ' , nx.minimum_node_cut(G_c))
print('Minimum edge cut: ' , nx.minimum_edge_cut(G_c))  
````

### Task (c) 

<font size="3">Find the minimum vertex coloring Z : V  N of G.</font>  

<font size="3">We use ```smallest_last``` because of the minimum number of colors</font>  

````python
node_colors = nx.greedy_color(G_c, 'smallest_last')

colors = np.array(list(node_colors.values()))
print('Number of colors:', len(np.unique(colors)))

plt.rcParams["figure.figsize"] = (20, 10)
colors = ['hotpink', 'dodgerblue', 'limegreen', 'yellow']
nx.draw_planar(G_c, node_color=[colors[node_colors[node]] for node in G_c.nodes()], with_labels = True, edge_color = 'skyblue', font_size = 12, node_size = 100)
````

### Task (d) 

 <font size="3">Find the minimum edge coloring X : E  N of G.</font>   

<font size="3">Which is the same as building a vertex coloring for the corresponding line graph and transferring resulting colors to vertices of the initial graph</font> 

````python
G_l = nx.line_graph(G_c)
edge_colors = nx.greedy_color(G_l, 'smallest_last')

#print(edge_colors.values())
print('Number of edge colors :', len(np.unique(np.array(list(edge_colors.values())))))

colors = ['hotpink', 'dodgerblue', 'limegreen', 'yellow', 'skyblue', 'olivedrab', 'orange', 
          'mediumpurple', 'lightsalmon']

nx.draw_planar(G_c, with_labels = True, edge_color = [colors[edge_colors[edge]] for edge in G_c.edges()], font_size = 12, node_size = 100, node_color = 'lightgray')
````

### Task (e)

<font size="3">Find the maximum clique Q  V of G.</font>

````python
from networkx.algorithms.approximation import clique

cliques = list(nx.find_cliques_recursive(G_c))

clique_sizes = np.array([len(el) for el in cliques])
print('Maximum clique size: ', np.max(clique_sizes))
print('Maximum clique:', clique.max_clique(G_c))
````

### Task (l)

<font size="3">Find all biconnected components (blocks) and draw the block-cut tree of G.</font>

````python
biconnected_components = list(nx.biconnected_components(G_c))
block_cut_tree = nx.Graph()

for vertex in nx.articulation_points(G_c): 
    for i in range(len(biconnected_components)):
        if vertex in biconnected_components[i]: block_cut_tree.add_edge(vertex, i + 1)

nx.draw_planar(block_cut_tree, with_labels = True, node_size = 400, font_size = 20, node_color = 'palevioletred', edge_color = 'skyblue')
print(biconnected_components)
````

### Task (m)

<font size="3">Find all 2-edge-connected components of G.</font>

````python
node_colors = {}
color = 0
for component in nx.k_edge_components(G_c, 2):
    for node in component:
        node_colors[node] = color
    color += 1 


colors = ['hotpink', 'dodgerblue', 'limegreen', 'yellow', 'skyblue', 'olivedrab', 'orange', 
          'mediumpurple', 'lightsalmon']

nx.draw_planar(G_c, node_color = [colors[node_colors[node]] for node in G_c.nodes()], with_labels = True, edge_color = 'skyblue')
````
